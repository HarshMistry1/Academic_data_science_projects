```python
import sys
import time
from getpass import getpass
import json


def Main():
    print('*'*78)
    print('*'*78)
    print()
    print('Welcome to Bank application'.center(80))
    print()
    print('*'*78)
    print('*'*78)
    print()
    print('1.Login \t\t2.Signup \t\t 3.Exit'.center(50))
    print()
    usr=int(input('Enter your choice: '))
    if usr==1:
        Login()
    elif usr==2:
        Signup()
    elif usr==3:
        sys.exit()
    else:
        print('Invalid choice: ')
        time.sleep(2)
        Main()
        
def Login():
    print('Welcome to Login page'.center(80))
    print()
    ac=input('Enter your account number: ')
    with open('Bal.json') as f:
        a = json.load(f)
        f.close()
    if ac in a.keys():
        pas=getpass('Enter your password: ')
        if a[ac]['password']==pas:
            print()
            print(f'Welcome, {a[ac]["fname"]} {a[ac]["lname"]}.')
            print()
            print('1.Debit \t 2.Credit \t 3.check balance \t 4. Change password \t 5. Logout')
            uch=int(input('Enter your choice: '))
            if uch==1:
                Debit(ac)
            elif uch==2:
                Credit(ac)
            elif uch==3:
                Check_bal(ac)
            elif uch==4:
                o=mail(ac)
                ot=input('Enter otp: ')
                if ot==o:
                    cpas=getpass('enter current password: ')
                    newpass=getpass('Enter new password: ')
                    con_pass=getpass('enter confirm password: ')
                    with open('Bal.json') as f:
                        a = json.load(f)
                        f.close()
                    if cpas==a[ac]['password'] and newpass==con_pass and cpas!=newpass:
                        a[ac]['password']=newpass
                        f = open('Bal.json', 'w')
                        json.dump(a,f, indent=2)
                        f.close()
                        print('Your password has changed successfully....')
                        time.sleep(3)
                        Login()
                    else:
                        print('something wrong.....')
                        time.sleep(2)
                        Login()             
                    
                else:
                    print('invalid otp')
            elif uch==5:
                print()
                print()
                print(f'You have signed out {a[ac]["fname"]}.')
                print()
                print()
                Main()
            else:
                print('invalid choice.....')
            
        else:
            print('invalid username or password')
            time.sleep(2)
            Login()
    
    
    else:
        print('no result found.')
        print('1.Try to log in again  \t 2. Sign Up')
        a=int(input('Enter the choice:'))
        if a==1: 
            time.sleep(2)
            Login()
        elif a==2:
            time.sleep(2)
            Signup()
        else:
            print("Please enter the correct choice.")
            Main()
            
def Home(ac):
    print()
    print('1.Debit \t 2.Credit \t 3.check balance \t 4. Change password \t 5. Logout')
    with open('Bal.json') as f:
        a = json.load(f)
        f.close()
    uch=int(input('Enter your choice: '))
    if uch==1:
        Debit(ac)
    elif uch==2:
        Credit(ac)
    elif uch==3:
        Check_bal(ac)
    elif uch==4:
        Chpass(ac)
    elif uch==5:
        print(f'You have signed out{a[ac]["fname"]}.')
        Main()
    else:
        print('invalid choice.....')
        Home(ac)
        
def Dir(ac):
    with open('Bal.json') as f:
        a = json.load(f)
        f.close()
    print("1.Home  \t 2.Log Out")
    choice = int(input('Enter your choice:'))
    if choice==1:
        Home(ac)
    elif choice==2:
        print()
        print()
        print(f'You have signed out {a[ac]["fname"]}.')
        print()
        print()
        Main()
    else:
        print("Please, fill the correct value.")
        Dir(ac)
    
        
def Debit(ac):
    time.sleep(2)
    amount=int(input('Enter amount: '))
    with open('Bal.json') as f:
        a = json.load(f)
        f.close()
    if a[ac]['bal']>=amount:
        a[ac]['bal']-=amount
        print(f'your amount {amount} has been debited')
        a[ac]['bal'] = a[ac]['bal']
        f = open('Bal.json', 'w')
        json.dump(a,f, indent=2)
        f.close()
        print(f'your updated balance {a[ac]["bal"]}')
        Dir(ac)
    else:
        print(f'Sorry, your current balance is:  {a[ac]["bal"]}.')
        time.sleep(1)
        Debit(ac)
        
def Credit(ac):
    time.sleep(2)
    amount=int(input('Enter amount: '))
    with open('Bal.json') as f:
            a = json.load(f)
            f.close()
    if amount!=0:
        a[ac]['bal']+=amount
        a[ac]['bal'] = a[ac]['bal']
        f = open('Bal.json','w')
        json.dump(a,f,indent=2)
        f.close()
        print(f'your amount {amount} has been Credited')
        print()
        print(f'your current balance is {a[ac]["bal"]}.')
        print()
        Dir(ac)
        

def Check_bal(ac):
    with open('Bal.json') as f:
        a = json.load(f)
        f.close()
    print()
    print(f'Your balance is {a[ac]["bal"]}.')
    Dir(ac)
    

def Chpass(ac):
    o=mail(ac)
    ot=input('Enter otp: ')
    if ot==o:
        cpas=getpass('enter current password: ')
        newpass=getpass('Enter new password: ')
        con_pass=getpass('enter confirm password: ')
        with open('Bal.json') as f:
            a = json.load(f)
            f.close()
        if cpas==a[ac]['password'] and newpass==con_pass and cpas!=newpass:
            a[ac]['password']=newpass
            f = open('Bal.json', 'w')
            json.dump(a,f, indent=2)
            f.close()
            print('Your password has changed successfully....')
            time.sleep(3)
            Login()
        else:
            print('something wrong.....')
            time.sleep(2)
            Chpass()                 
    else:
        print('invalid otp')

    
def Signup():
    fname=input('Enter first name: ')
    lname=input('Enter last name: ')
    email=input('enter email accept only gmail id....: ')
    age=int(input('enter your age: '))
    ph_no=input('enter phone number: ')
    pas=getpass('enter password: ')
    c_pas=getpass('enter re passwoerd: ')
    if pas==c_pas:
        bal=1000
        with open('Bal.json') as f:
            a = json.load(f)
            f.close
        b = a.keys()
        c = list(b)
        c = c[-1]
        d = str(int(c)+1)
        o=otp()
        import smtplib
        con=smtplib.SMTP('smtp.gmail.com',587)
        con.starttls()
        con.login('bankapplication7@gmail.com','Bankapp@')
        con.sendmail('bankapplication7@gmail.com',email,f'Subject: OTP \n\n {o}')
        con.close()
        time.sleep(1)
        print('..')
        time.sleep(1)
        print('..')
        time.sleep(1)
        print('..')
        time.sleep(1)
        print()
        ot=input('Enter otp: ')
        while True:
            if ot==o:
                a[d]={'fname':fname,'lname':lname,'password':pas,'bal':bal,'email':email,'age':age}
                f = open('Bal.json','w')
                json.dump(a,f, indent=2)
                f.close()
                print('your account created....')
                import smtplib
                con=smtplib.SMTP('smtp.gmail.com',587)
                con.starttls()
                con.login('bankapplication7@gmail.com','Bankapp@')
                con.sendmail('bankapplication7@gmail.com',email,f'Subject: account \n\n your account is {d}')
                con.close()
                break
                Login()
            else:
                ot=input('Enter otp')
                continue
                
def otp():
    import random
    o=''
    p='01234567789'
    for i in range(6):
        o=o+random.choice(p)
    return o

def mail(ac):
    p=otp()
    with open('Bal.json') as f:
        a = json.load(f)
        f.close()
    import smtplib
    con=smtplib.SMTP('smtp.gmail.com',587)
    con.starttls()
    con.login('bankapplication7@gmail.com','Bankapp@')
    con.sendmail('bankapplication7@gmail.com',f'{a[ac]["email"]}',f'Subject: OTP \n\n {p}')
    con.close()
    return p
        
Main()
```
